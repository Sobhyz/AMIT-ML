```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../Data science/CarPrice_Assignment.csv")
```

```python
df
```

# Drop Dublicates

```python
df = df.drop_duplicates()
```

```python
df.describe()
```

Car Id is irrelevant to the price

```python
df.drop('car_ID',axis=1,inplace=True)
```

```python
df['CarName'] = df['CarName'].apply(lambda x : x.split()[0])
```

```python
df['CarName'].unique()
```

there are some spelling mistakes

```python
df['CarName'].replace('maxda','mazda',inplace = True)
df['CarName'].replace('Nissan','nissan',inplace = True)
df['CarName'].replace('porcshce','porsche',inplace = True)
df['CarName'].replace('toyouta','toyota',inplace = True)
df['CarName'].replace(['vokswagen','volkswagen'],'vw',inplace = True)
```

```python
df.isnull().sum()/len(df.index)
```

Data is Clean

```python
categorical = list(df.columns[df.dtypes == 'object'])
categorical
```

```python
plt.figure(figsize=(20,20))
for i in categorical:
    plt.subplot(3,4,categorical.index(i)+1)
    sns.barplot(x=i,y='price',data=df)
plt.show()
```
