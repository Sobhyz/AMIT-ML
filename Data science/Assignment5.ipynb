```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(1,51,1)
plt.plot(x,x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
y = np.arange(50,0,-1)
plt.plot(x,x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,x,label = 'first line', color = 'r', linewidth = 5)
plt.plot(x,y,label = 'first line', color = 'g', linewidth = 2)
plt.legend()
plt.show()
```

```python
plt.plot(x,x, 'r--',label = 'first line', linewidth = 5)
plt.plot(x,y, 'g*-',label = 'first line', linewidth = 2)
plt.legend()
plt.show()
```

```python
x_new = [1,4,6,8]
y_new = [5,20,1,8]
plt.plot(x_new,y_new,'ro--')
plt.show()
```

```python
PL = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
Popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.bar(x=PL,height=Popularity,width=0.3)
plt.show()
```

```python
x_pos = [i for i, j in enumerate(PL)]
plt.barh(x_pos,Popularity)
plt.yticks(x_pos, PL)
plt.show()
```

```python
men = (22, 30, 35, 35, 26)
women = (25, 32, 30, 35, 29)
x = np.array([1,2,3,4,5])
width=0.3
plt.bar(x=x,height=men,label = 'men', width=width)
plt.bar(x=x+width,height=women,label = 'women', width=width)
plt.legend()
```

```python
plt.pie(Popularity,labels=PL,autopct='%1.1f%%')
```
